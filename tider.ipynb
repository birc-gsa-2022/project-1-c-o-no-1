```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 10
M = 10

#size = N*M

m1 = 16
m2 = 32
m3 = 64


n = np.array([1000, 5000, 10000, 15000, 20000, 25000, 30000, 40000, 45000, 50000, 75000, 100000])
#m = np.array([15, 15, 30, 30, 50, 50, 100, 100, 100, 200])

naive_time_1 = np.array([0.265, 0.107, 0.311, 0.672, 1.147, 1.746, 2.481, 4.323, 5.463, 6.682, 14.878, 26.193])
naive_time_2 = np.array([0.014, 0.106, 0.327, 0.671, 1.145, 1.753, 2.471, 4.324, 5.431, 6.708, 15.003, 26.181])
naive_time_3 = np.array([0.015, 0.107, 0.327, 0.671, 1.147, 1.745, 2.479, 4.325, 5.443, 6.692, 14.812, 26.167])

lin_time_1 = np.array([0.011, 0.015, 0.023, 0.029, 0.035, 0.037, 0.033, 0.054, 0.061, 0.064, 0.085, 0.111])
lin_time_2 = np.array([0.007, 0.016, 0.022, 0.029, 0.034, 0.039, 0.032, 0.054, 0.058, 0.052, 0.084, 0.110])
lin_time_3 = np.array([0.007, 0.014, 0.021, 0.028, 0.031, 0.036, 0.032, 0.055, 0.044, 0.063, 0.081, 0.108])
```

```python
tnm_naive_1 = naive_time_1/(m1*n)
tnm_naive_2 = naive_time_2/(m2*n)
tnm_naive_3 = naive_time_3/(m3*n)

tnm_lin_1 = lin_time_1/(m1+n)
tnm_lin_2 = lin_time_2/(m2+n)
tnm_lin_3 = lin_time_3/(m3+n)
```

```python
fig, ax = plt.subplots()

ax.plot(n, naive_time_1, label='Naive, m=16', marker='.', color='c', linewidth=0.5)
ax.plot(n, naive_time_2, label='Naive, m=32', marker='.', color='m', linewidth=0.5)
ax.plot(n, naive_time_3, label='Naive, m=64', marker='.', color='y', linewidth=0.5)

ax.plot(n, lin_time_1, label='Linear, m=16', marker='*', color='c', linewidth=0.5)
ax.plot(n, lin_time_2, label='Linear, m=32', marker='*', color='m', linewidth=0.5)
ax.plot(n, lin_time_3, label='Linear, m=64', marker='*', color='y', linewidth=0.5)


ax.legend()
plt.xlabel("Length of string, n")
#plt.ylabel("Time/nm")
plt.title("Comparison of times of naive and linear algorithm")
```

It can be seen that the linear implementation is pretty much linear. Unfortunately, it looks like the naive implementation is quadratic in *n*, rather than in *n* and *m*, and the cause is unknown.
